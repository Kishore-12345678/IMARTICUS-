```python
sns.countplot(data=df, x='RAM')
plt.title('Mobile Processors')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.countplot(x='RAM', hue='Launched Price (India)', data=df)
plt.xticks(rotation=45)
plt.title('Price Range per Processor Type')
```

```python
from sklearn.preprocessing import LabelEncoder

# Make a copy to keep original data safe
df_encoded = df.copy()

# Initialize LabelEncoder
le = LabelEncoder()

# Convert processor column from strings to numbers
df_encoded['RAM'] = le.fit_transform(df_encoded['RAM'])

# Now it's safe to use in ML models or numerical analysis
print(df_encoded[['RAM', 'Launched Price (India)']].head())
```
